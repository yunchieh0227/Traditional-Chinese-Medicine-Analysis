##資料視覺化

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 資料夾路徑
csv_folder_path = '/content/drive/MyDrive/PointAugment/data_csv/white'
png_folder_path = '/content/drive/MyDrive/PointAugment/data/png/white'

# 獲取所有的 .csv 檔案
csv_files = [f for f in os.listdir(csv_folder_path) if f.endswith('.csv')]
count = 0
for csv_file in csv_files:
    if count >= 4:
      break
    # 讀取 .csv 檔案
    data = pd.read_csv(os.path.join(csv_folder_path, csv_file))

    # 第一行之後的第一列代表時間值
    time_values = data.iloc[1:, 0].values.astype(float)

    # 第一列之後的第一行代表波長值
    wavelength_values = data.columns[1:].astype(float)

    # 其餘的數據代表強度值
    intensity_values = data.iloc[1:, 1:].values.astype(float)

    # 將時間和波長的陣列重塑為與強度陣列相符的形狀
    time_values_reshaped = np.repeat(time_values[:, np.newaxis], wavelength_values.shape[0], axis=1)
    wavelength_values_reshaped = np.repeat(wavelength_values[np.newaxis, :], time_values.shape[0], axis=0)

    # 將陣列壓平以進行散點圖繪製
    time_values_flattened = time_values_reshaped.flatten()
    wavelength_values_flattened = wavelength_values_reshaped.flatten()
    intensity_values_flattened = intensity_values.flatten()

    # 建立一個圖形
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    # 建立一個色彩映射
    cmap = plt.cm.get_cmap('RdBu_r')

    # 獲取最小和最大波長值進行色彩縮放
    wavelength_min = np.min(wavelength_values_flattened)
    wavelength_max = np.max(wavelength_values_flattened)

    # 根據波長創建一個3D散點圖，顏色基於波長
    scatter = ax.scatter(time_values_flattened, wavelength_values_flattened, intensity_values_flattened,
                         c=(wavelength_values_flattened - wavelength_min) / (wavelength_max - wavelength_min), cmap=cmap)

    # 設定軸的標籤
    ax.set_xlabel('Time')
    ax.set_ylabel('Wavelength')
    ax.set_zlabel('Intensity')

    # 改變視角
    ax.view_init(elev=20, azim=225)  # 修改這些值以改變視角

    # 增加一個顏色條
    fig.colorbar(scatter, ax=ax, label='Wavelength')

    # 儲存圖片
    plt.savefig(os.path.join(png_folder_path, csv_file.replace('.csv', '.png')))

    # 關閉圖片，避免在記憶體中累積
    plt.close()
    count = count + 1
```
